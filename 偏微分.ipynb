```python
def func(x0, x1):
    return x0**2 + x1**2
```

```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x0 = np.arange(-3, 3, 0.25)
x1 = np.arange(-3, 3, 0.25)
x0, x1 = np.meshgrid(x0, x1)
y = func(x0, x1)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x0, x1, y)
plt.show()
```
